```python
import music21
import numpy as np
score1 = music21.converter.parse('xml/Jos2604.xml')
score2 = music21.converter.parse('humdrum/Jos_all/Jos2604-Deus_pacis_reduxit.krn')
```

```python
data1 = []
for i in score1.flat:
    if isinstance(i, music21.note.Note):
        data1.append((i.ps, i.quarterLength))

data2 = []
for i in score2.flat:
    if isinstance(i, music21.note.Note):
        data2.append((i.ps, i.quarterLength))

for i, j in zip(data1, data2):
    if i != j:
        print i, j
```
